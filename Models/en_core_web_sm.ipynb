```python
##Load model
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe("spacytextblob")
```

```python
##Read in data
df = pd.read_csv('x_en_dataset.csv')
df.head()
```

```python
##Calculating polarity score
df['calculated_sentiment'] = df['cleaned_text'].apply(lambda text: nlp(text)._.blob.polarity)
```

```python
##Review outputs
df.head()
```

```python
##Finding Metrics
##Need to change continous variables to integers
def to_int_class(x):
    return max(1, min(5, round(((x + 1) * 2) + 1)))

y_true = df['sentiment'].apply(to_int_class)
y_pred = df['calculated_sentiment'].apply(to_int_class)

acc = accuracy_score(y_true, y_pred)
print("Accuracy:", acc)

f1 = f1_score(y_true, y_pred, average='macro')
print("F1 Score:", f1)
```
